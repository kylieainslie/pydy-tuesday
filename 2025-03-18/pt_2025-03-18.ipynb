# Palm tree sizes
Let's look at how the sizes of different species of palms vary across subfamilies. First, let's just make a table of mean and sd of palm tree size by subfamily. The variables we'll need are `palm_subfamily` and `max_stem_height_m`.

```python
import numpy as np

# remove NA values
df = df[['palm_subfamily', 'max_stem_height_m', 'main_fruit_colors']].dropna()

# Summary stats by subfamily
size_by_subfam = df.groupby('palm_subfamily')['max_stem_height_m'].agg(['mean', 'std', 'min','max'])

print(size_by_subfam)
```

We see from the table of summary statistics that **Calamoideae** has the highest mean and maximum stem height, but also the most variability among the species in this subfamily. The summary statistics for the maximum stem heights of **Arecoideae** and **Coryphoideae** are fairly similar. However, let's investigate the distribution of maximum stem heights among species in each subfamily in a bit more detail.

We will use a violin plot to look at the distribution of maximum height by subfamily using a violin plot. We will exclude subfamily **Nypoideae** because it doesn't contain enough information.

Looking at the violin plot, we can see that most of the species within the **Arecoideae** subfamily cluster towards the bottom of the height distribution. We see a similar pattern in the **Coryphoidae** subfamily, but it's less dramatic than in **Arecoideae**. As suggested by the summary statistics, the **Calamoideae** subfamily has the most variability in maximum stem height and the species within this subfamily have somewhat more evenly distributed stem heights at the lower end of the range, but with some species that can have very tall stem heights.

```python
import plotnine as p9

# Remove subfamily Nypoideae
df = df[df["palm_subfamily"] != "Nypoideae"]

# Create violin plot
plot = (
    p9.ggplot(df, p9.aes(x="palm_subfamily", y="max_stem_height_m", fill="palm_subfamily"))
    + p9.geom_violin(alpha=0.7)
    + p9.theme_minimal()
    + p9.labs(
        title="Palm Tree Heights by Subfamily",
        x="Subfamily",
        y="Maximum Height (m)"
    )
    + p9.theme(
        axis_text_x=p9.element_text(rotation=45, hjust=1),
        legend_position='none'
    ) 
)

plot
```

Now, let's see if there's any relationship (visually) to maximum stem height and fruit color, by plotting the different species by there fruit color on the violin plot. Since some species have fruit with more than one color, we'll use the first listed. Also, for ease of visibility, we're group similar colors into categories. For example, "white", "cream", and "ivory" will be grouped as "white".

```python
# Split the 'main_fruit_colors' by the semicolon to separate colors
split_fruit_colors = df['main_fruit_colors'].str.split(';', expand=True)

# Create new vars for separated fruit colors
df['first_color'] = split_fruit_colors[0]  # Always exists
df['second_color'] = split_fruit_colors[1] if split_fruit_colors.shape[1] > 1 else None  # Assign None if there's no second color

# Define a mapping from individual fruit colors to broader categories
color_mapping = {
    'black': 'Black',  
    'brown': 'Brown',
    'green': 'Green',  
    'yellow': 'Yellow',   
    'red': 'Red',  
    'orange': 'Orange',  
    'purple': 'Blue',  
    'white': 'White',  
    'pink': 'Red', 
    'straw-coloured': 'Yellow', 
    'ivory': 'White',  
    'cream': 'White',  
    'grey': 'Black',  
    'blue': 'Blue'
}

category_colors = {
    'Red': '#FF6347',    # Tomato Red
    'Orange': '#FFA500', # Orange
    'Yellow': '#FFD700', # Gold
    'Green': '#008000',  # Green
    'Blue': '#1E90FF',   # Dodger Blue
    'Black': '#000000',  # Black
    'White': '#FFFFFF',  # White
    'Brown': '#8B4513',  # Saddle Brown
}

# Add a new column with the mapped category
df['color_category'] = df['first_color'].map(color_mapping)

# Summarize how many species of each subfamily have fruit of each color category
df_summary = df.groupby(['palm_subfamily', 'color_category']).size().reset_index(name='species_count')

plot_summary = (
    p9.ggplot(df_summary, p9.aes(x='palm_subfamily', y='species_count', fill='color_category'))
    + p9.geom_bar(stat='identity', position='dodge', show_legend=True, alpha=0.6)
    + p9.scale_fill_manual(values=category_colors) 
    + p9.theme_minimal()
    + p9.labs(
        title="Fruit Color Distribution",
        x="Count",
        y="Palm Subfamily"
    )
    + p9.theme(
        axis_text_x=p9.element_text(rotation=45, hjust=1),
        axis_text_y=p9.element_text(size=10),
        legend_position='bottom'
    )
    + p9.guides(
        fill=p9.guide_legend(title="Fruit Color")
    )
)

plot_summary
```

```python
# Create violin plot with jittered points
plot2 = (
    p9.ggplot(df, p9.aes(x="palm_subfamily", y="max_stem_height_m", fill="palm_subfamily"))
    + p9.geom_violin(alpha=0.7)
    + p9.geom_jitter(p9.aes(color='color_category'), width=0.1, size=1, alpha=0.6) 
    + p9.scale_color_manual(values=category_colors) 
    + p9.theme_minimal()
    + p9.labs(
        title="Palm Tree Heights by Subfamily",
        x="Subfamily",
        y="Maximum Height (m)"
    )
    + p9.theme(
        axis_text_x=p9.element_text(rotation=45, hjust=1),
        legend_position='bottom'
    ) 
    + p9.guides(
        fill='none',
        color=p9.guide_legend(title="Fruit Color")
    )
)

plot2
```
